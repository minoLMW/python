````python
# 'ğŸƒ'ë¥¼ ì´ìš©í•˜ì—¬ ì•„ë˜ì™€ ê°™ì€ ë„í˜•ì„ ë§Œë“¤ì–´ë³´ì.
'''
ğŸƒ ğŸƒ ğŸƒ ğŸƒ ğŸƒ
ğŸƒ ğŸƒ ğŸƒ ğŸƒ ğŸƒ
ğŸƒ ğŸƒ ğŸƒ ğŸƒ ğŸƒ
ğŸƒ ğŸƒ ğŸƒ ğŸƒ ğŸƒ
ğŸƒ ğŸƒ ğŸƒ ğŸƒ ğŸƒ
'''
for i in range(5):
    for j in range(5):
        print('ğŸƒ', end=' ')
    print()

'''
ğŸƒ ğŸƒ ğŸƒ ğŸƒ ğŸƒ
ğŸƒ ğŸƒ ğŸƒ ğŸƒ
ğŸƒ ğŸƒ ğŸƒ
ğŸƒ ğŸƒ
ğŸƒ
'''
for i in range(5):
    for j in range(i, 5): # range(0, 5), range(0, 1)
        print('ğŸƒ', end=' ')
    print()

'''
ğŸƒ
ğŸƒ ğŸƒ
ğŸƒ ğŸƒ ğŸƒ
ğŸƒ ğŸƒ ğŸƒ ğŸƒ
ğŸƒ ğŸƒ ğŸƒ ğŸƒ ğŸƒ
'''
for i in range(5):
    for j in range(0, i+1):
        print('ğŸƒ', end=' ')
    print()
````
